```python
# Gráfico de linha (Line Plot)

import pandas as pd
import matplotlib.pyplot as plt

# Criando um DataFrame com dados de vendas
data = {
    'Mes': ['Jan', 'Fev', 'Mar', 'Abr', 'Mai'],
    'Vendas': [1000, 1200, 900, 1300, 1100]
}

df = pd.DataFrame(data)

# Definindo a coluna 'Mes' como índice
df.set_index('Mes', inplace=True)

# Plotando gráfico de linha
df.plot(kind='line', marker='o', title='Vendas por Mês')
plt.ylabel('Vendas')
plt.xlabel('Mês')
plt.grid()
plt.show()
```

```python
# Gráfico de barras (Bar Plot)

df.plot(kind='bar', title='Vendas por Mês', color='green')
plt.ylabel('Vendas')
plt.xlabel('Mês')
plt.show()
```

```python
# Histograma (Hist Plot)

import numpy as np

df_hist = pd.DataFrame({'Notas': np.random.normal(7, 1.5, 1000)})

df_hist.plot(kind='hist', bins=30, title='Distribuição de Notas', color='purple')
plt.xlabel('Nota')
plt.show()
```

```python
# Gráfico de dispersão (Scatter Plot)

df_scatter = pd.DataFrame({'Altura': [1.50, 1.60, 1.70, 1.80, 1.90], 
                           'Peso': [50, 60, 70, 80, 90]})

df_scatter.plot(kind='scatter', x='Altura', y='Peso', title='Altura vs Peso', color='red')
plt.show()
```

```python
# Importar bibliotecas
from sklearn.linear_model import LinearRegression
import numpy as np

# Dados fictícios: tamanho da casa (m²) e preço (em milhares de reais)
X = np.array([[50], [70], [90], [120], [150]])  # Tamanho (entrada)
y = np.array([200, 280, 350, 470, 600])        # Preço (saída)

# Criar e treinar o modelo
modelo = LinearRegression()
modelo.fit(X, y)

# Fazer uma previsão para uma casa de 100 m²
tamanho_novo = np.array([[100]])
preco_previsto = modelo.predict(tamanho_novo)
print(f"Preço previsto para uma casa de 100 m²: R$ {preco_previsto[0]:.2f} mil")

# Avaliar o modelo (coeficiente de determinação R²)
score = modelo.score(X, y)
print(f"Acurácia do modelo (R²): {score:.2f}")
```
