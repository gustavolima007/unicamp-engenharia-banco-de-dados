## Exercicios de Agrupamento

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
import numpy as np

# Gerando dados aleatórios
np.random.seed(42)
data = np.random.rand(10, 2)

# Aplicando o agrupamento hierárquico
Z = linkage(data, 'ward')

# Plotando o dendrograma
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

data = np.array([[1, 100], [2, 200], [3, 300]])

scaler = StandardScaler()
normalized_data = scaler.fit_transform(data)
print(normalized_data)
```

```python
from sklearn.cluster import KMeans
import numpy as np

data = np.array([[1, 2], [1, 4], [1, 0],
                 [4, 2], [4, 4], [4, 0]])

kmeans = KMeans(n_clusters=2, random_state=0).fit(data)
print(kmeans.labels_)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(100, 2)
sse = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0).fit(data)
    sse.append(kmeans.inertia_)

plt.plot(k_range, sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np
import time

# Dados grandes
X = np.random.rand(10000, 2)

# K-means
start = time.time()
kmeans = KMeans(n_clusters=3).fit(X)
print("Tempo de execução:", time.time() - start)
```

```python
from sklearn.cluster import KMeans
import numpy as np

X = np.array([[1, 2], [1, 4], [1, 0], [4, 2]])
kmeans = KMeans(n_clusters=2, tol=1e-4)  # Tolerância para parada
kmeans.fit(X)
print("Número de iterações:", kmeans.n_iter_)
```

```python
from sklearn.cluster import KMeans
import numpy as np

X = np.array([[1, 2], [1, 4], [1, 0], [4, 2]])
kmeans_plus = KMeans(n_clusters=2, init='k-means++').fit(X)
print("Centroides:", kmeans_plus.cluster_centers_)
```

```python
import numpy as np

# Dois pontos
p1 = np.array([1, 2])
p2 = np.array([4, 6])

# Distância Euclidiana
dist = np.sqrt(np.sum((p1 - p2) ** 2))
print("Distância Euclidiana:", dist)
```
